```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data = pd.read_csv("admision.csv")
data.head()
```

```python
# Check for missing values
data.isnull().sum().sum()
```

```python
# Get dataset information
data.info()
```

```python
# Mean of each column
print('Mean of each column')
data.mean()
```

```python
# Maximum GRE Score
data['GRE Score'].max()
```

```python
# Minimum value of each column
print('Minimum value of each column')
data.min()
```

```python
# Most frequent CGPA value (Mode)
data['CGPA'].mode()[0]
```

```python
# Mode of all columns
print('Mode for each column')
data.mode().iloc[0]
```

```python
# Range of each column (Max - Min)
print('Range of each column')
data.max() - data.min(),
```

```python
# Standard deviation of each column
print('Standard Deviation of each column')
data.std()
```

```python
# Variance of each column
print('Variance of each column')
data.var(),
```

```python
# Median (50th percentile) of each column
print('50th Percentile (Median) : ')
data.median()
```

```python
#describe of data
data.describe()
```

```python
# Histogram for each feature
data.drop(columns=['Serial No.']).hist(bins=10, figsize=(15, 10), color='burlywood', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
# KDE plots for each feature
plt.figure(figsize=(15, 15))
for i, column in enumerate(data.columns, 1):
    plt.subplot(3, 3, i)  # Adjust grid size if needed
    sns.kdeplot(data[column], color='DARKORANGE')  # Change color here
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
# Histogram with KDE overlay for each feature
plt.figure(figsize=(15, 15))
for i, column in enumerate(data.columns, 1):
    plt.subplot(3, 3, i)  # Adjust grid size if needed
    sns.histplot(data[column], bins=10, color='indianred', kde=True, edgecolor='black')
    plt.title(f'Distribution of {column}')
```

```python
# Overall histogram visualization
data.hist(figsize = (16,9))
```

```python
# Separate data based on Research participation
research_yes = data[data['Research'] == 1]  # Students who did research
research_no = data[data['Research'] == 0]  # Students who didn't do research


# GRE Score distribution based on Research participation
plt.figure(figsize=(10, 6))
sns.histplot(research_yes['GRE Score'], bins=20, color='royalblue', label='Did Research', kde=True, stat="frequency")
sns.histplot(research_no['GRE Score'], bins=20, color='tomato', label='Did Not Do Research', kde=True, stat="frequency")

plt.legend()
plt.show()



# Add labels and title
plt.title('GRE Score Distribution Based on Research Participation', fontsize=16)
plt.xlabel('GRE Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
